# About the dataset
This dataset is valuable for various stakeholders such as real estate agents, property managers, researchers, and policymakers to understand trends in housing markets, evaluate property investments, or assess rental affordability





	Posted On: date when the property is listed for availability to rent
	BHK: count of Bedroom-Hall-Kitchen
	Rent: Total Amount to rent the Property
	Size: size of the property in Square feet
	Floor: the floor where the property is located
	Area Type: area the listed property located at
	Area Locality: Address of the listed property
	City: top 6 cities in india
	Furnishing Status: is the list property furnished, semi-furnished,  Unfurnished
	Tenant Preferred: Bachelor or Family
	Bathroom: number of Bathroom
	Point of Contact: to contact owner or to contact agent

# Feature Engineering

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
import category_encoders as ce
```

```python
rent_final = rent_id.drop(columns='Posted On')
```

```python
#rent_final = rent_id.drop(columns="Rent")
```

```python
#split into X and y
rent_final
```

```python
x = rent_final.drop(columns='Rent')
y = rent_final['Rent'] #target
```

```python
y
```

```python
x.info()
#independent Variable
```

# Encoding with LeaveOneOutEncoder

```python
encoder = ce.LeaveOneOutEncoder()
x = encoder.fit_transform(x,y)
```

# Splitting Data for Training and Testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

# Instanciating the Regressor models

```python
model_rfr = RandomForestRegressor(n_estimators=7)
model_dt = DecisionTreeRegressor()
model_lr = LinearRegression()
```

```python
models = [model_rfr, model_dt, model_lr]
```

# Fitting the Models to Traing data

```python
for model in models:
    print("fitting model:",(model))
    model.fit(x_train,y_train)
```

# Score of Training data

```python
for model in models:
    print("score of", (model),"for training data:", (model.score(x_train,y_train)))
```

# Score of testing data

```python
for model in models:
    print("score of", (model),"for testing data:", (model.score(x_test,y_test)))
```

# Creating Function for regression Results

```python
def regression_results(y_test, y_predicted):
    
    mean_absolute_error = metrics.mean_absolute_error(y_test, y_predicted)
    mse = metrics.mean_squared_error(y_test, y_predicted)
    r2 = metrics.r2_score(y_test, y_predicted)

    print('r2:', round(r2, 4))
    print('MAE:', round(mean_absolute_error,4))
    print('MSE:', round(mse, 4))
    print('RMSE:', round(np.sqrt(mse), 4))
```

# Predicting values for X_test 

```python
for model in models[:]:
    y_predicted = model.predict(x_test)
    
    print(f"Report:{model}")
    print(f"{regression_results(y_test, y_predicted)}")
    
    #compare = pd.DataFrame({"Actual":y_test,"Predicted":y_predicted})
    
    #compare1 = compare.head(40)
    #compare1.plot(kind='bar',figsize=(16,10))
    #plt.grid(which='major',linestyle='-',linewidth='0.5',color='green')
    #plt.grid(which='minor',linestyle=':',linewidth='0.5',color='black')
    #plt.show()
```
